```python
# merge functionality- we perform merge basically by looking at column values
```

```python
data1= pd.DataFrame({"key1": [1,2,4,5,6],
                     "key2": [4,5,6,7,8],
                     "key3": [3,4,5,6,6]})
```

```python
data2= pd.DataFrame({"key1": [1,2,45,6,67],
                     "key4": [56,5,6,7,8],
                     "key5": [3,56,5,6,6]})
```

```python
data1
```

```python
data2
```

```python
pd.merge(data1, data2)     # by default it does inner merge
```

```python
# inner merge- it works in this way, key1 is the common column name, so the values that are common in both data1 and data2
# dataframe's column key 1 are taken. Corresponding to those values that row record is taken and displayed.
# by default the how parameter of merge() function= inner.
# we can also give how= left, right, cross, outer etc.
```

```python
pd.merge(data1, data2, how="left")
```

```python
# left merge- data1 is left merged to data2. So, all the columns of data1 are taken. 
# Then those columns of data2 are taken which match i.e. values of key1 column of data2 which are same in key1 column of data1, its corresponding row record is taken. 
# NaN is displayed where its not matching
```

```python
pd.merge(data1, data2, how="right")
```

```python
# right merge- data2 is right merged to data1. So, all the columns of data2 are taken. 
# Then the columns of data1 that match i.e values of key1 column of data1 that match to key1 column of data2, its corresponding row records are taken. 
# NaN is displayed when its not matching.
```

```python
pd.merge(data1, data2, how="outer")
```

```python
# outer merge- All the values of key1 column of both data1 and daat2 are taken(unique values). 
# Corresponding to those values, those row records are taken for key2, key3(corresponding to key1 of data1) 
# and key4, kay5(corresponding to key1 of data2). 
# NaN is displayed when its not matching.
```

```python
pd.merge(data1, data2, how="cross")
```

```python
# cross merge- It takes one record of data1 and combines it with all the records of data2. 
# It continues this combination for each record of data1.
```

```python
pd.merge(data1, data2, how="left", on="key1")
```

```python
pd.merge(data1, data2, how="left", left_on="key2", right_on="key4")
```

```python
# window function- 
```

```python
df9= pd.DataFrame({"a": [1,2,3,4,5,6,7,8,9]})
```

```python
df9
```

```python
df9.rolling(window=1).mean()
```

```python
# window=1 means it takes just one data and find its mean i.e. mean of 1= 1, mean of 2= 2 etc. 
# rolling just helps in going from one data to next.
```

```python
df9.rolling(window=2).mean()
```

```python
# window=2 means it takes two datas at once and find their mean.
# for the 1st data, it considers the previous data to be NaN. So, (NaN+1)/2 = NaN. 
# now rolling lands us at data= 2. So, mean= (1+2)/2= 1.5 and so on
```

```python
df9.rolling(window=3).mean()
```

```python
# window=3 means it takes three datas at once and find their mean.
# for the 1st data= 1, it considers the previous 2 data to be NaN. So, (NaN+NaN+1)/3 = NaN. 
# now rolling lands us at data= 2. So, mean= (NaN+1+2)/3= NaN
# now rolling lands us at data=3. so mean= (1+2+3)/3= 2 and so on
```

```python
df9.rolling(window=3).sum()
```

```python
# it takes 3 data at once as window=3. For data= 1, the prev two data are NaN, so sum= NaN+NaN+1= NaN
# rolling lands us at data= 2, the prev two data re NaN and 1, so sum= NaN+1+2 = NaN
# now rolling lands us at data= 3, the prev two data are 1, 2 so sum= 1+2+3= 6 and so on
```

```python
df9.rolling(window=3).min()    # it will find the minimum of 3 data at once (window=3)
```

```python
df9.cumsum()    # this is the cumulative sum
```

```python
# here window size is not taken, so the sum is cumulative. For 1st data= 1, sum= 1, for 2nd data=2, sum= 1+2= 3,
# for 3rd data= 3, sum= 3+3= 6, for 4th data= 4, sum= 6+4= 10 and so on
```
