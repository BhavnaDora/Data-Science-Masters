```python
# beautifying it a bit-
soup= BeautifulSoup(response.content, "html.parser")
```

```python
soup.find_all("img")
```

```python
# we get a list with all the img tags which contains the images at the src link. The first element is not an elon musk image so we delete it.
# on clicking the src link we land on webpage containing that image.
# so we have to extract all these image links.
```

```python
image_tags= soup.find_all("img")
```

```python
del image_tags[0]
```

```python
image_tags
```

```python
len(image_tags)            # will give how many elements i.e. images present in the list
```

```python
image_data_mongo= []                # create an empty list, this will be pushed later on to MondoDB

for i in image_tags:
    image_url= i['src']                                 # extract the link and store it in image_url variable
    image_data= requests.get(image_url).content         # all the data from that image's url has been scraped and stored in a variable image_data
    mydict={"index": image_url, "image": image_data}     # we create a dictionary which stores the image url as well as the data for each image
    image_data_mongo.append(mydict)
    with open (os.path.join(save_dir, f"{query}_{image_tags.index(i)}.jpg"), "wb") as f:
        f.write(image_data)
```
