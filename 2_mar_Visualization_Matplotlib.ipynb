#### Visualization- Matplotlib

```python
# To derive sentiments i.e. perform some sort of sentiment analysis on a huge dataset, 
# we need visualization tools like matplotlib, seaborn, plotly etc.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.random.rand(50)       # 1-D array will be generated with 50 random numbers
```

```python
y= np.random.rand(50)       # 1-D array will be generated with 50 random numbers
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)          # a scatter plot with points which are x and y pairs is obtained
```

```python
# we can beautify this plot by labelling-

plt.scatter(x,y) 
plt.xlabel("this is my x axis")
plt.ylabel("this is my y axis")
plt.title("this is my x vs y")
```

```python
# we can also give different colours to the points by using the c parameter os scatter plot.
# we can also choose a hexadecimal colour code by searching in google- HTML color codes, copy that code and give it in parameter c

plt.scatter(x,y, c="r")
plt.xlabel("this is my x axis")
plt.ylabel("this is my y axis")
plt.title("this is my x vs y")
```

```python
plt.scatter(x,y, c="#522C46")
plt.xlabel("this is my x axis")
plt.ylabel("this is my y axis")
plt.title("this is my x vs y")
```

```python
# we can also control the figure size too-
# we can also choose intensity of points using the parameter alpha of scatter plot-
# we can also represent the points as X instead of dots using the parameter marker of scatter plot-
# we can use grid in the plot-

plt.figure(figsize= (6,2))
plt.scatter(x,y, c="#522C46", alpha= 0.6, marker= "X")
plt.xlabel("this is my x axis")
plt.ylabel("this is my y axis")
plt.title("this is my x vs y")
plt.grid()
```

```python
x= np.linspace(1,10,100)     # it will generate 100 equally spaced points between 1 to 10
```

```python
y= np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)          # a line plot is obtained which is a sinusoidal curve
```

```python
plt.scatter(x,y)              # scatter plot is obtained which also resembles sine curve
```

```python
plt.plot(x,y, "--r")
```

```python
plt.figure(figsize= (8,2))
plt.plot(x,y)
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("sine wave graph")
plt.grid()
```

```python
# plotting categorical vs. numerical data using bar plot-
```

```python
x= ["a", "b", "c", "d", "e"]
y= np.random.rand(5)      # 1-D array with 5 random numbers is generated
```

```python
x
```

```python
y
```

```python
# to plot each element of x against each element of y we do a bar plot-

plt.bar(x,y)
```

```python
# some beautification of the plot-

plt.figure(figsize= (5,3))
plt.bar(x,y, color= "#EF8F05")
plt.xlabel("representing my categorical data")
plt.ylabel("representing my numerical data")
plt.title("this is my bar plot")
plt.grid()
```

```python
# to represent the bar plot horizontally-

plt.figure(figsize= (5,3))
plt.barh(x,y, color= "#EF8F05")
plt.title("this is my horizontal bar plot")
plt.grid()
```

```python
x= [3,4,5,6,7,8]
y= [4,5,6,7,8,9]
```

```python
plt.figure(figsize=(6,2))
plt.plot(x,y)       # this will give a line plot
```

```python
plt.scatter(x,y)           # this will give scatter plot i.e pair of-each element of x corresponding to each element of y
```

```python
# we want to plot a data on x axis with its element's frequency on y axis-

data= [1,2,3,4,5,6,5,6,6,7,7,8,8,8,8,8]
```

```python
plt.figure(figsize= (6,3))
plt.hist(data)        # histogram plot shows the frequency of occurance of data points on y axis
```

```python
# beuatifying it-

plt.figure(figsize= (6,3))
plt.hist(data)       
plt.xlabel("representing values")
plt.ylabel("representing frequency")
plt.title("histogram plot")
plt.grid()
plt.savefig("plot1.png")             #we can save this plot too
```

```python
# Plotting 3-D graph-
```

```python
x= np.random.rand(50)
y= np.random.rand(50)
z= np.random.rand(50)
```

```python
fig= plt.figure()
ax= fig.add_subplot(projection= "3d")
ax.scatter(x,y,z, c="#AF4C61", alpha=0.9)
plt.xlabel("my x axis")
plt.ylabel("my y axis")
plt.show()
```
