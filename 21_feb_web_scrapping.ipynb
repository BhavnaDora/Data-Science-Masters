#### Web scrapping

```python
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
import logging
```

```python
# We will scrap information from flipkart website. 
# In the flipkart website millions of products are listed and each of them has reviews.
# Our objective is to acquire all these data.
```

```python
flipkart_url= "https://www.flipkart.com/search?q=" + "tv"       # we can use this for different different keywords q, so we are able to customise the url
```

```python
flipkart_url
```

```python
# on clicking the above url we land in flipkart's website where the keyword tv has already been entered into the search bar and search results are already displayed.
# This is the work of Web API. 
# Many products are listed, and on clicking them we are directed to that product's page where reviews are present, our objective is to get those
```

```python
# Instead of clicking on the link to open the url, we can do-
urlopen(flipkart_url)
```

```python
urlclient= urlopen(flipkart_url)              #storing this into a variable urlclient
```

```python
urlclient.read()            # it will read the entire webpage(all the html codes, search results etc)
```

```python
flipkart_page= urlclient.read()     # storing all these data in the variable flipkart_page
```

```python
# to make this data a bit more readable and structured we use BeautifulSoup- 
flipkart_html= bs(flipkart_page, 'html.parser')
```

```python
flipkart_html
```

```python
# now we will fetch all the urls of all the products displayed as search results-
# let's start with the hyperlink of one product-
# we did F12 on the flipkart web page and used inspect element cursor, then we observe that all the data are divided into boxes. Their html codes are shown on the right hand side-
# we expand those html codes going by div classes till we reach the hyperlink of the product and we copy it

"https://www.flipkart.com" + "/samsung-80-cm-32-inch-hd-ready-led-smart-tizen-tv-2022-model/p/itm3c07ddc2dd25d?pid=TVSGHY2UZA9YHWQN&lid=LSTTVSGHY2UZA9YHWQNNQENTU&marketplace=FLIPKART&q=tv&store=ckf%2Fczl&spotlightTagId=BestsellerId_ckf%2Fczl&srno=s_1_3&otracker=search&fm=organic&iid=5dd49456-5dda-4570-aa37-0db0f00161c2.TVSGHY2UZA9YHWQN.SEARCH&ppt=sp&ppn=sp&ssid=oekmiow7680000001677691255981&qH=c9a1fdac6e082dd8"
```

```python
# we see that for each product, we have the hyperlink in div class(html code)-

flipkart_html.find_all("div", {"class":"_1AtVbE col-12-12"})       #it will fetch those whose marker is div and class name is _1AtVbE col-12-12 
```

```python
# so we got a list with all the links.
bigbox= flipkart_html.find_all("div", {"class": "_1AtVbE col-12-12"})
```

```python
len(bigbox)     # returns how many elements are present in that list i.e. how many links for products
```

```python
del bigbox[0:4]
```

```python
# let us extract the html link of first product-
# we start from div class where the class name is "_1AtVbE col-12-12" and reach till a class which has the link-

bigbox[0].div.div.div.a["href"]
```

```python
"https://www.flipkart.com" + bigbox[0].div.div.div.a["href"]
```

```python
# for getting the links of all the products-

for i in bigbox:
    print("https://www.flipkart.com" + i.div.div.div.a["href"])
```

```python
# consider a single product-

productlink= "https://www.flipkart.com" + bigbox[0].div.div.div.a["href"]
```

```python
productlink
```

```python
# we want to extract all the written information of the product page i.e. ratings, comment headers, entire comment, customer's name etc.

requests.get(productlink)     # this scraps every information we see on the product page
```

```python
product_req= requests.get(productlink)
```

```python
product_req.text
```

```python
# to make this dump of information a bit more readable we use BeautifulSoap-

product_html= bs(product_req.text, "html.parser")
```

```python
product_html
```

```python
# we open the product page and inspect it using F12. We see that the comments are present in separate boxes. 
# On expanding we can see the ratings, comment headers, comment, person's name.
# All these we have already scraped and stored in product_html.
# Now we have to extract these
```

```python
product_html.find_all("div", {"class": "_16PBlm"})       # this class name corresponds to all the comment boxes inside reviews
                                                         # so we extract all these big comment boxes first
```

```python
comment_box= product_html.find_all("div", {"class": "_16PBlm"})             # the above list obtained is stored in the variable comment_box
```

```python
len(comment_box)      
```

```python
# let's fetch the person's name-

comment_box[0].div.div.find_all("p", {"class": "_2sc7ZR _2V5EHH"})
```

```python
comment_box[0].div.div.find_all("p", {"class": "_2sc7ZR _2V5EHH"})[0].text
```

```python
# fetching all people's names-

for i in comment_box:
    print(i.div.div.find_all("p", {"class":"_2sc7ZR _2V5EHH"})[0].text)
```

```python
# fetch the ratings-
# let's first fetch the first comment's rating-

comment_box[0].div.div.find_all("div", {"class": "_3LWZlK _1BLPMq"})
```

```python
comment_box[0].div.div.find_all("div", {"class": "_3LWZlK _1BLPMq"})[0].text
```

```python
# fetching all the ratings-

for i in comment_box:
    print(i.div.div.find_all("div", {"class": "_3LWZlK _1BLPMq"})[0].text)
```

```python
# fetch the comment headers-

comment_box[0].div.div.find_all("p", {"class": "_2-N8zT"})
```

```python
# fetching all comment headers-

for i in comment_box:
    print(i.div.div.find_all("p", {"class": "_2-N8zT"})[0].text)
```
