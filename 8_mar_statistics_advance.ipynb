#### Quantile-Quantile(QQ) plot-

```python
# check if normal distribution or not using QQ plot-
```

```python
import scipy.stats as stat
import numpy as np
import seaborn as sns
import pylab
```

```python
df= sns.load_dataset("iris")
df.head()
```

```python
# check whether each feature follows normal distribution or not-
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_data(df, feature):
    plt.figure(figsize=(8,6))
    plt.subplot(1,2,1)
    sns.histplot(df[feature], kde=True)
    plt.subplot(1,2,2)
    stat.probplot(df[feature], dist="norm", plot=pylab)
    plt.show()
```

```python
plot_data(df, "sepal_width")
```

```python
# in the probability plot i.e. QQ plot, the ordered values in the y axis are basically the feature of the dataframe i.e.
# sepal_width here. The theoritical quantiles on x axis are coming due to the dist="norm". On plotting these two, 
# if most points lie on the line, then we can say the feature of the dataframe i.e. sepal_width follows a normal distribution.
```

```python
plot_data(df, "sepal_length")
```

```python
# sepal_length follows a gaussian distribution because we can see in the QQ plot that most points lie on the line.
```

```python
plot_data(df, "petal_length")
```

```python
# petal_length doesn't follow a guassian distribution except for one segment.
```

```python
plot_data(df, "petal_width")
```

```python
# one part of the data follows gaussian distribution but not the entire data of the feature petal_width.
```

```python
df1= sns.load_dataset("tips")
df1.head()
```

```python
plot_data(df1, "total_bill")
```

```python
# the feature total_bill follows a log normal distribution rather than a gaussian distribution because it is right skewed.
```

```python
plot_data(df1, "tip")
```

```python
# tip feature also foloows a log normal distribution because it is right skewed.
```
