#### We do feature engineering and EDA on Date_of_Journey column-

```python
df["Date"]= df["Date_of_Journey"].str.split("/").str[0]
df["Month"]= df["Date_of_Journey"].str.split("/").str[1]
df["Year"]= df["Date_of_Journey"].str.split("/").str[2]
```

```python
df.info()
```

```python
# the 3 new columns- Date, Month, Year have object datatypes so we typecast it to int-
df["Date"]= df["Date"].astype(int)
df["Month"]= df["Month"].astype(int)
df["Year"]= df["Year"].astype(int)
```

```python
df.info()
```

```python
# now we drop Date_of_Journey column- 
df.drop("Date_of_Journey", axis=1, inplace=True)
```

```python
df["Arrival_Time"]= df["Arrival_Time"].str.split(" ").str[0]

df["Arrival_hour"]= df["Arrival_Time"].str.split(":").str[0]
df["Arrival_min"]= df["Arrival_Time"].str.split(":").str[1]
```

```python
# the 2 new columns- Arrival_hour and Arrival_min have object datatypes, so we typecast them to int-
df["Arrival_hour"]= df["Arrival_hour"].astype(int)
df["Arrival_min"]= df["Arrival_min"].astype(int)
df.info()
```

```python
# now we drop the Arrival_Time column-
df.drop("Arrival_Time", axis=1, inplace=True)
```

#### Now we do feature engineering & EDA for Total_Stops column-

```python
df["Total_Stops"].unique()
```

```python
df["Total_Stops"].isnull().sum()
```

```python
df[df["Total_Stops"].isnull()]
```

```python
df["Total_Stops"].mode()
```

```python
# We replace the NaN value with mode of the column, we encode them using ordinal encoding concept, assign non-stop as 0, 1 stop as 1, 2 stops as 2 and so on, 
# this is because greater the no. of stops, greater is the price.

df["Total_Stops"]= df["Total_Stops"].map({"non-stop": 0, "1 stop": 1, "2 stops": 2, "3 stops": 3, "4 stops": 4, np.nan: 1})
```

#### Now we do feature engineering & EDA on these 4 categorical features-

```python
# we perform one hot encoding-
from sklearn.preprocessing import OneHotEncoder
encoder= OneHotEncoder()
encoded= encoder.fit_transform(df[["Airline", "Source", "Destination", "Additional_Info"]]).toarray()
encoded
```

```python
encoded_df= pd.DataFrame(encoded, columns= encoder.get_feature_names_out())
encoded_df
```

```python
# we then drop the 4 categorical featueres from df -
df.drop(["Airline", "Source", "Destination", "Additional_Info"], axis=1, inplace=True)
```

```python
# then we concatenate the encoded_df dataframe with df and we feed that data frame to the model for training.
```

```python
df_final= pd.concat([df, encoded_df], axis=1)
df_final
```
