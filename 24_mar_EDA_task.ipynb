```python
df["Classes"]= df["Classes"].str.strip()
```

```python
df["Classes"].unique()
```

```python
df["Classes"].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder
ord_encoder= OrdinalEncoder(categories=[["not fire", "fire"]])
encoded= ord_encoder.fit_transform(df[["Classes"]])
encoded
```

```python
encoded_df= pd.DataFrame(encoded, columns=["Classes_encoded"])
encoded_df
```

```python
df.drop("Classes", axis=1, inplace=True)
df= pd.concat([df, encoded_df], axis=1)
df
```

#### Insights/Observation-

```python
sns.catplot(x="Classes_encoded", y="Temperature", data=df, kind="box")
```

#### Insights/observation-
- The temperature is lower for class 0 i.e. no forest fire, but the temperature is higher for class 1 i.e. forest fire

```python
import seaborn as sns
sns.countplot(x="month", data= df[df["region"]==1], hue="Classes_encoded")
```

#### Insights/observations-
- For the region= 1 i.e Bejaia region, the above count plot shows the no. of forest fires and not forest fires in different months.
- Month 8 has the highest count of forest fires.
